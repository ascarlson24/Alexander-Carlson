```python
Sales = float(input("Enter Sales ($): "))
print(f"Sales: $", Sales)
Costs_of_goods_sold = 0.4 * Sales
print(f"Costs of goods sold: $", Costs_of_goods_sold)
Gross_Profit = Sales - Costs_of_goods_sold
print(f"Gross Profit: $" , Gross_Profit)

```

design simple numerical calculator in python. user provides 3 inputs: two numbers and mathematical operator. program should print output

```python
N1 = (float(input("Numer 1: ")))
N2 = (float(input("Numer 2: ")))
M = ((input("Mathematical Operator: ")))

if M == "+":
  print(N1 + N2)
elif M == "-":
  print(N1 - N2)
elif M == "*":
  print(N1 * N2)
elif M == "/":
  print(N1 / N2)



```
