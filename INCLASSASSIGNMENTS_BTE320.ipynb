```python
Sales = float(input("Enter Sales ($): "))
print(f"Sales: $", Sales)
Costs_of_goods_sold = 0.4 * Sales
print(f"Costs of goods sold: $", Costs_of_goods_sold)
Gross_Profit = Sales - Costs_of_goods_sold
print(f"Gross Profit: $" , Gross_Profit)
tax_rate = float(input("Enter tax rate (%): "))
taxes = Gross_Profit * tax_rate / 100
print(f"Taxes: $", taxes)
net = Gross_Profit - taxes
print(f"Net Income: $", net)

```

design simple numerical calculator in python. user provides 3 inputs: two numbers and mathematical operator. program should print output

```python
N1 = (float(input("Numer 1: ")))
N2 = (float(input("Numer 2: ")))
M = ((input("Mathematical Operator: ")))
res = "None"

if M == "+":
  res = N1 + N2
elif M == "-":
  res = N1 - N2
elif M == "*":
  res = N1 * N2
elif M == "/":
  res = N1 / N2
elif M == "//":
  res = N1 // N2
elif M == "**":
  res = N1 ** N2
elif M == "%":
  res = N1 % N2
else:
  print("Invalid Operation")
print(f"Result:", res)
```

```python
import math

print("Ad\tProfit")
tp = 10
standard = 20

for ad in range(0, 225, 25):
  extra = 2 * round(math.sqrt(ad))
  revenue = (extra * tp) + (standard * tp)
  profit = revenue - (200 + ad)
  print(f"${float(ad)}\t${float(profit)}")
```

```python
print("Hours     Fee")
hours = 1
while hours < 9:
  total = 5 + hours * 2.5
  while total < 10:
    total = 10.0
  while total > 20:
    total = 20.0
  print(f"{hours}        ${total}")
  print()
  hours += 1
```

```python
import math

print("Ad\tProfit")
tp = 10
standard = 20
ad = 0
while ad < 225:
  extra = 2 * round(math.sqrt(ad))
  revenue = (extra * tp) + (standard * tp)
  profit = revenue - (200 + ad)
  print(f"${float(ad)}\t${float(profit)}")
  ad += 25
```

```python
import random #
answer = input("Do you want to play (y/n) ?: ")

while answer == "y":
  choice = input("choose [rock , paper , scissors]: ").lower()
  computer_move = random.choice(["rock", "paper", "scissors"])
  print()
  print(f"Player: {choice}")
  print(f"Computer: {computer_move}")

  if choice == computer_move:
    print("It's a tie!")
  elif (
    (choice == "scissors" and computer_move == "paper") or
    (choice == "rock" and computer_move == "scissors") or
    (choice == "paper" and computer_move == "rock")
    ):
    print("You Win!")
  else:
    print("You Lose!")
  print()
  answer = input("Do you want to play again (y/n) ?: ")
```
